#Integer Location

##iloc is used to read and display specific rows

To print out the number of rows we want with integer as the row identifier

```python
ds.iloc[0]
```

We can also make a new DataFrame with iloc[ [ ] ]

```python
ds.iloc[[0,1,2]]  #Prints 3 rows with index -0,1,2
```

##Specific Column for specific Rows

###Using iloc we can print out specific column for a specified row index

iloc[[r1,r2,....,rn],[c1,c2,...cn]] - to read the specific column value to the corresponding row value. We must only use integers for all the specifying values in the "[ ]"

```python
ds.iloc[[1,3,4],[2,4,6,3,1]]
```

```python
ds.iloc[[1],[1,2,3]]
```

###Loc- Label(name) based Read and display

Using loc[ ] we can read and print the values using the column labels.

Syntax - ds.loc[[r1,r2,....rn],[c_name1,c_name2,....c_name_n]]

```python
ds.loc[[0,1],['Open','Close']]
```

Individual Value for a specific plot

```python
ds.loc[1,"Open"]
```

```python
ds.loc[[1,2,3],"Open"]
```
